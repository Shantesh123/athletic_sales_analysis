```python
# Import Libraries and Dependencies
import pandas as pd
import numpy as np
from pathlib import Path
from datetime import datetime 
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')

```

```python
# Display the 2020 sales DataFrame

athletic_sales_2020 = athletic_sales_2020.dropna(how="any")
athletic_sales_2020.head(100)
```

```python
# Display the 2021 sales DataFrame

athletic_sales_2021 = athletic_sales_2021.dropna(how="any")
athletic_sales_2021.head()
```

```python
# Check the 2020 sales data types.
athletic_sales_2020.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
#athletic_sales_2020_2021 = athletic_sales_2020.join(athletic_sales_2021,lsuffix='_2020',rsuffix='_2021')
athletic_sales_2020_2021=pd.concat([athletic_sales_2020,athletic_sales_2021],axis=0,join='inner').reset_index(drop=True)
athletic_sales_2020_2021.head()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
def custom_sum(x):
    return x.sum()

num_products_country_state_city = athletic_sales_2020_2021.groupby(['region', 'state','city']).apply(lambda x: pd.Series({"Total_Products_Sold": custom_sum(x["units_sold"])}))

# Show the top 5 results.
num_products_country_state_city.head(20)                        
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_products_country_state_city = pd.pivot_table(athletic_sales_2020_2021, 
                                     index=['region','state','city'],
                                     values='units_sold',
                                     aggfunc='sum')
pivot_products_country_state_city.head(20)
# Rename the "units_sold" column to "Total_Products_Sold"

pivot_products_country_state_city = pivot_products_country_state_city.rename(columns={"units_sold": "Total_Products_Sold"})
# Show the top 5 results.
pivot_products_country_state_city.head(20)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
athletic_sales_2020_2021.drop_duplicates()
pivot_product_retailer_country_state_city = pd.pivot_table(athletic_sales_2020_2021, 
                                     index=['product','retailer','region','state','city'],
                                     values='total_sales',
                                     aggfunc='sum')
pivot_product_retailer_country_state_city = pivot_product_retailer_country_state_city.pivot_table(index="retailer", 
                                                 columns="product",
                                                 values="total_sales").rename_axis(None, axis=1)

renamed_results=pd.DataFrame(pivot_product_retailer_country_state_city['Women\'s Athletic Footwear'].sort_values(ignore_index=True))
renamed_results
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
retailer_units_country_state_city = athletic_sales_2020_2021.groupby(['product','retailer','region', 'state','city']).apply(lambda x: pd.Series({"Total_Units_Sold": custom_sum(x["units_sold"])}))

# Show the top 5 results.
#retailer_units_country_state_city
result = retailer_units_country_state_city.loc['Women\'s Athletic Footwear', :]
renamed_results=pd.DataFrame(result.rename({'Total_Units_Sold': 'Womens_Footwear_Units_Sold'}, axis='columns'))
renamed_results.head(5)
# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_units_product_retailer_country_state_city = pd.pivot_table(athletic_sales_2020_2021, 
                                     index=['product','retailer','region','state','city'],
                                     values='units_sold',
                                     aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
result_units = retailer_units_country_state_city.loc['Women\'s Athletic Footwear', :]
# Show the top 5 results.
renamed_results=pd.DataFrame(result_units.rename({'Total_Units_Sold': 'Womens_Footwear_Units_Sold'}, axis='columns'))
renamed_results.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_invoice_retailer_country_state_city = pd.pivot_table(athletic_sales_2020_2021, 
                                     index=['product','invoice_date'],
                                     values='total_sales',
                                     aggfunc='sum')
pivot_invoice_retailer_country_state_city = pivot_invoice_retailer_country_state_city.pivot_table(index="invoice_date", 
                                                 columns="product",
                                                 values="total_sales").rename_axis(None,axis=1)
day_most_women_ath_sales=pivot_invoice_retailer_country_state_city.sort_values(by=['Women\'s Athletic Footwear'], ascending=False)
sales_data_df = pd.DataFrame(day_most_women_ath_sales.iloc[:,4])
top_day= pd.DataFrame(sales_data_df.head(1))
top_day
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
sales_data_df.resample('D').sum().sort_values(by=['Women\'s Athletic Footwear'], ascending=False)

# Sort the resampled pivot table in ascending order on "Total Sales".
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

sales_data_df.resample('W').sum().sort_values(by=['Women\'s Athletic Footwear'], ascending=False)
# Sort the resampled pivot table in ascending order on "Total Sales".
```
